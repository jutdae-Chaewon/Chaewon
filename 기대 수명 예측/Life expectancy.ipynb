### **기대 수명 예측**

- 다중선형회귀 이용
- 특성 공학, 릿지, 라쏘 기법 사용
- 데이터를 시각화해서 각 특성과 기대 수명 사이의 관계를 어떻게 설계할 것인지 고민해보기
- test -> (330,19)

```python
import pandas as pd

df = pd.read_csv("train.csv")
df
```

```python
df.columns.tolist()
```

```python
x = df[['Adult Mortality', 'Alcohol',
        'infant deaths' , 'under-five deaths ', ' HIV/AIDS',
        ' BMI ' , ' thinness  1-19 years', ' thinness 5-9 years',
        'GDP','percentage expenditure','Income composition of resources',
        'Polio', 'Diphtheria ','Hepatitis B'
        ]]
y = df[["Life expectancy "]]

x = x.to_numpy()
y = y.to_numpy()

print(x.shape)
print(y.shape)

#1,4 x
#1,3 x
#1,5 x
#1,2 x
#5 (6)
#4 (7)
#3 (6)
#'Measles ' -> 6
#'Measles ', ' HIV/AIDS' -> x
#'under-five deaths ' -> 7
# ' BMI ' -> 7
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias=False, degree=4)  # 5제곱항까지 생성 가능

poly.fit(x_train)
x_train_poly = poly.transform(x_train)
print(x_train_poly.shape)
```

```python
poly.fit(x_test)  
x_test_poly = poly.transform(x_test)
print(x_test_poly.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x_train_poly, y_train)

print("학습용 데이터셋", lr.score(x_train_poly, y_train))
print("시험용 데이터셋", lr.score(x_test_poly, y_test))
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(x_train_poly)

x_train_scaled = ss.transform(x_train_poly)
x_test_scaled = ss.transform(x_test_poly)
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge()

ridge.fit(x_train_scaled, y_train)

print("학습용 데이터셋: ", ridge.score(x_train_scaled, y_train))
print("시험용 데이터셋: ", ridge.score(x_test_scaled, y_test))
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []

alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]

for a in alpha_list:
    ridge = Ridge(alpha=a)
    ridge.fit(x_train_scaled, y_train)
    
    train_score.append(ridge.score(x_train_scaled, y_train))
    test_score.append(ridge.score(x_test_scaled, y_test))
    
print(train_score)
print(test_score)
```

```python
ridge = Ridge(alpha=100)
ridge.fit(x_train_scaled, y_train)

print("학습용 데이터셋: ", ridge.score(x_train_scaled, y_train))
print("시험용 데이터셋: ", ridge.score(x_test_scaled, y_test))
```

```python
from sklearn.metrics import mean_absolute_error

y_pred1 = ridge.predict(x_test_scaled)

mae = mean_absolute_error(y_test, y_pred1)
print(mae)
```
